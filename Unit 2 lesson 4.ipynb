```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('c:/charcoal_production.csv')
df
```

```python
columns = ['Year', 'Quantity']
df1 = pd.DataFrame(df, columns=columns)
df1
```

```python
plt.figure()

a = df1['Year']
b = df1['Quantity']

plt.ylabel('''Production in Metric Tons (x1000)''')
plt.xlabel('Year')
plt.ylim([400, 1200])
plt.title('Charcoal Energy Production in USA 1990-2014')

plt.plot(a,b)

```
