```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
binomial = np.random.binomial(1, .3, 100)
plt.hist(binomial)
plt.axvline(binomial.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(binomial.mean() + binomial.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(binomial.mean() - binomial.std(), color='b', linestyle='dashed', linewidth=2)
plt.show()
# Descriptive statistics not useful here as they do not give any real meaning related 
# to the central tendency of the data here since we only have 2 values. The central tendency
# is not important. 
```

```python
chisquare = np.random.chisquare(.5, 100)
plt.hist(chisquare)
plt.axvline(chisquare.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(chisquare.mean() + chisquare.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(chisquare.mean() - chisquare.std(), color='b', linestyle='dashed', linewidth=2)
plt.show()

# I think the descriptive statistics do provide useful information about the variable here. 
# the mean acuratle describe the central tendency of the data. The onyl problem here would
# be in describing the variance since the 1st negative deviation doesnt contain any of the 
# variables since the majority of the variables are concentrated so close to the mean. 
```

```python
laplace = np.random.laplace(3, 1, 1000)
plt.hist(laplace)
plt.axvline(laplace.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(laplace.mean() + laplace.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(laplace.mean() - laplace.std(), color='b', linestyle='dashed', linewidth=2)
plt.show()

# The descriptive statistics give good information about the central tendency and 
# variance as the data has a relatively normal distribute and the standard deviations 
# and mean are populated with data on all sides. 
```

```python
pareto = np.random.pareto(1, 100)
plt.hist(pareto)
plt.axvline(pareto.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(pareto.mean() + pareto.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(pareto.mean() - pareto.std(), color='b', linestyle='dashed', linewidth=2)
plt.show()

# Descriptive statistics do not give good measures of the data. Variables are not 
# distributed evently and the stanard deviations do not give us much insight 
# about the data
```

```python
rayleigh = np.random.rayleigh(1, 100)
plt.hist(rayleigh)
plt.axvline(rayleigh.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(rayleigh.mean() + rayleigh.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(rayleigh.mean() - rayleigh.std(), color='b', linestyle='dashed', linewidth=2)
plt.show()

# The descriptive statistics do give us ueful infmation aboutt he variable as it is 
# generally pretty evenly distributed and we can derive insights from this data
```

```python
uniform = np.random.uniform(1, 5, 1000)
plt.hist(uniform)
plt.axvline(uniform.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(uniform.mean() + uniform.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(uniform.mean() - uniform.std(), color='b', linestyle='dashed', linewidth=2)
plt.show()

# The descriptive statistics do not seem to provide useful data as the data seems to be 
# spread out consistenly with no real central tendency annd no significance
# to the standard deviations. 
```

```python
additional = np.random.normal(5, .05, 100)
additional_1 = np.random.normal(10, 2, 100)
new = additional + additional_1
plt.hist(new)
plt.axvline(new.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(new.mean() + new.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(new.mean() - new.std(), color='b', linestyle='dashed', linewidth=2)
plt.show() 
# I think the descriptive statistics here give meaninful insights into the central 
# tendency and variability of variables. The data is evenly spread out 
# and the mean/standard deviations contain a good deal of the data. 
```
